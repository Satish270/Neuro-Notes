```python
import numpy as np
import csv
import pandas as pd

num_samples = 10000  # or even more for a large dataset
num_elements = 8
d = 0.5
wavelength = 1.0

def steering_vector(num_elements, d, wavelength, theta_deg):
    theta = np.radians(theta_deg)
    return np.exp(1j * 2 * np.pi * d * np.arange(num_elements) * np.sin(theta) / wavelength)

with open('nsb_dataset.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(['desired_angle', 'interf_angle1', 'interf_angle2'] + [f'w{i}_real' for i in range(num_elements)] + [f'w{i}_imag' for i in range(num_elements)])
    for _ in range(num_samples):
        desired = np.random.uniform(-90, 90)
        interf1 = np.random.uniform(-90, 90)
        interf2 = np.random.uniform(-90, 90)
        # Avoid overlap
        while abs(desired-interf1)<5 or abs(desired-interf2)<5 or abs(interf1-interf2)<5:
            interf1 = np.random.uniform(-90, 90)
            interf2 = np.random.uniform(-90, 90)
        v_desired = steering_vector(num_elements, d, wavelength, desired)
        V_interf = np.vstack([steering_vector(num_elements, d, wavelength, angle) for angle in [interf1, interf2]]).T
        P = np.eye(num_elements) - V_interf @ np.linalg.pinv(V_interf)
        w = P @ v_desired
        w = w / np.linalg.norm(w)
        row = [desired, interf1, interf2] + list(np.real(w)) + list(np.imag(w))
        writer.writerow(row)

# Read the generated CSV file and display the first few rows
df = pd.read_csv('nsb_dataset.csv')
print(df.head())
```
